# Analisis de los precios de casas usadas en la Region Metropolitana, Chile 18/07/2023

```python
# hacemos import de las librerias
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
# importamos los datos
df = pd.read_csv(r"C:\Users\danie\OneDrive\Escritorio\RoadMap Data Analys\Data Analyst Portfolio Project\Precios Casas RM.csv")
df
```

```python
# eliminamos duplicados
df = df.drop_duplicates()
df
```

```python
# vemos que valores tenemos en los datos
df.info()
```

```python
df.describe()
```

```python
# ahora vemos si tenemos valores nulos
df.isnull().sum()
```

```python
# eliminamos los nulos ya que no hay manera de recuperar esos datos
df = df.dropna()
df
```

```python
# eliminaremos columnas que no nos interesan
df = df.drop(["id", "Realtor", "Price_CLP", "Price_USD", "Ubicacion"], axis=1)
df
```

```python
# veamos si existe alguna correlacion
corr = df.corr(numeric_only=True)
corr
```

```python
# hacemos un plot de la correlacion
colores = sns.light_palette("salmon", as_cmap=True)
mask = np.triu(corr)

with sns.axes_style("white"):
    plt.figure(figsize=(11,6))
    sns.heatmap(corr, cmap=colores, mask=mask, square=True, annot=True, fmt=".2f")
```

```python
# ordenamos por precio uf mas alto
df = df.sort_values("Price_UF", ascending=False)
df
```

```python
df.tail(10)
```

```python
# eliminamos valoes 0 en Price_UF
df_filtrado = df[df["Price_UF"] != 0]
df_filtrado
```

```python
df_filtrado.describe()
```

```python
# cantidad de habitaciones por casa
dormitoriosPerCasa = df_filtrado["Dorms"].value_counts().head(10)

x = dormitoriosPerCasa.index
y = dormitoriosPerCasa.values
x = x.sort_values()

# grafico
plt.bar( x, y, color="salmon")
plt.title("Cantidad de dormitorios por casas")
plt.xlabel("Numero de dormitorios")
plt.ylabel("Numero de csas")

plt.show()
```

```python
 plt2 = df_filtrado.groupby('Comuna')['Price_UF'] \
    .agg(['mean', 'count']) \
    .query('count >= 250') \
    .sort_values('mean')['mean'] \
    .plot(kind = 'barh', figsize= (15,5), title= 'Cantidad de avisos por comuna y sus precios promedio', color="salmon")
plt2.set_xlabel('Precio en UF')
plt2.set_ylabel('Comuna')
```

```python
plt3 = df_filtrado.groupby("Comuna")["Dorms"]\
    .agg(["mean", "count"])\
        .query("count >= 250")\
            .sort_values("mean")["mean"]\
                .plot(kind = 'barh', figsize= (15,5), color="salmon", title= 'Promedio de habitaciones por casa, por comuna')
plt3.set_xlabel('Cantidad de habitaciones promedio')
plt3.set_ylabel('Comuna')
```
